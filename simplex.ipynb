```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
from mpl_toolkits.mplot3d import Axes3D
 
fig = plt.figure(figsize=(10,10))
axes = fig.gca(projection='3d')
axes.view_init(-10, 10)
 
def f(x, y):
    return (15*x**2 + 25*y**2+3)
 
y = x = np.linspace(-3, 3, 50)
 
x, y = np.meshgrid(x, y)
z = f(x,y)
 

optimize.minimize(f, -3, args=(3))
optimization = optimize.minimize(f, -3, args=(3))
print(optimization.x)
 
surf = axes.plot_surface(x, y, z, cmap='coolwarm',
                       linewidth=0, antialiased=False)
print(optimization)
surf1 = axes.scatter(optimization.x, optimization.hess_inv, f(optimization.x,optimization.hess_inv), cmap='coolwarm',  s=200, marker='o')
 
axes.set_xlabel(' X')
axes.set_ylabel(' Y')
axes.set_zlabel(' Z')
 
plt.show()
```
